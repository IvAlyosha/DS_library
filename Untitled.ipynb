```python
def evaluate_preds(train_true_values, train_pred_values, test_true_values, test_pred_values):
    print(f"Train R2:{str(round(r2(train_true_values, train_pred_values), 3))}")
    print(f"Test R2:{str(round(r2(test_true_values, test_pred_values), 3))}")
    
    plt.figure(figsize=(18,10))
    
    plt.subplot(121)
    sns.scatterplot(x=train_pred_values, y=train_true_values)
    plt.xlabel('Предсказанные значения')
    plt.ylabel('Истинные значения')
    plt.title('Прогнозирование тренировочная выборка')
    
    plt.subplot(122)
    sns.scatterplot(x=test_pred_values, y=test_true_values)
    plt.xlabel('Предсказанные значения')
    plt.ylabel('Истинные значения')
    plt.title('Прогнозирование тестовая выборка')

    plt.show()
```

```python
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

Класс для обработки

```python
class DataPreprocessing:
    """Подготовка исходных данных"""

    def __init__(self):
        """Параметры класса"""
        self.medians = None
        self.kitchen_square_quantile = None
        
    def fit(self, X):
        """Сохранение статистик"""       
        # Расчет медиан
        self.medians = X.median()
        self.means = X.mean()
        self.mean_square_room = X.groupby("Rooms")['Square'].mean().reset_index()\
                        .rename(columns={'index':'Rooms', 'Square':'Mean_square'})
    
    def transform(self, X):
        """Трансформация данных"""

        # Rooms
        X['Rooms_outlier'] = 0
        X.loc[(X['Rooms'] == 0) | (X['Rooms'] >= 6), 'Rooms_outlier'] = 1
        X.loc[(X['Rooms'] == 0) & (X['Square'] <= 56.63), 'Rooms'] = 1
        X.loc[(X['Rooms'] == 0) & (X['Square'] <= 79.7) & (X['Square'] >= 56.63), 'Rooms'] = 2
        X.loc[(X['Rooms'] == 0) & (X['Square'] <= 116.49) & (X['Square'] >= 79.7), 'Rooms'] = 3
        X.loc[(X['Rooms'] == 0) & (X['Square'] <= 159.80) & (X['Square'] >= 116.49), 'Rooms'] = 4
        X.loc[(X['Rooms'] == 0) & (X['Square'] <= 243.04) & (X['Square'] >= 159.80), 'Rooms'] = 5
        X.loc[(X['Rooms'] > 6) & (X['Square'] <= 56.63), 'Rooms'] = 1
        X.loc[(X['Rooms'] > 6) & (X['Square'] <= 79.7) & (X['Square'] >= 56.63), 'Rooms'] = 2
        X.loc[(X['Rooms'] > 6) & (X['Square'] <= 116.49) & (X['Square'] >= 79.7), 'Rooms'] = 3
        X.loc[(X['Rooms'] > 6) & (X['Square'] <= 159.80) & (X['Square'] >= 116.49), 'Rooms'] = 4
        X.loc[(X['Rooms'] > 6) & (X['Square'] <= 243.04) & (X['Square'] >= 159.80), 'Rooms'] = 5
        
        # Square
        X = X.merge(self.mean_square_room, on='Rooms', how='left')
        X.loc[(X['Square'] < 15), 'Square'] = X['Mean_square'] 
        X.loc[(X['Square'] < 20) & (X['Rooms'] > 1 ) , 'Square'] = X['Mean_square']
        X.loc[(X['Square'] > 100) & (X['Rooms'] == 1 ) , 'Square'] = X['Mean_square']
        X.loc[(X['Square'] > 100) & (X['Rooms'] == 2 ) , 'Square'] = X['Mean_square']
        X.loc[(X['Square'] > 200) & (X['Rooms'] == 3 ) , 'Square'] = X['Mean_square']
        X.drop('Mean_square', axis=1, inplace=True)
        
        # KitchenSquare, LifeSquare
        X.loc[(X['KitchenSquare']>50), 'KitchenSquare'] = self.means['KitchenSquare']
        X.loc[(X['KitchenSquare'] > X['Square']),'KitchenSquare'] = self.means['KitchenSquare']
        X.loc[abs(X['KitchenSquare'] - X['Square'])<3, 'KitchenSquare'] = self.means['KitchenSquare']
        
        X.loc[(X['LifeSquare'].isna()) | (X['LifeSquare']<10), 'LifeSquare'] = X['Square'] * 0.65
        X.loc[(X['LifeSquare'] > X['Square']) & (X['Rooms'] == 1) & (X['KitchenSquare'] > 1),'LifeSquare'] = X['Square'] * 0.65
        X.loc[(X['LifeSquare'] > X['Square']) & (X['Rooms'] > 1),'LifeSquare'] = X['Square'] * 0.65
        X.loc[(X['LifeSquare'] > X['Square']) & (X['Rooms'] == 1),'LifeSquare'] = X['Square']
        X.loc[(X['LifeSquare'] == X['Square']), 'LifeSquare'] = X['Square'] * 0.65
        
        # HouseFloor, Floor
        X['HouseFloor_outlier'] = 0
        X.loc[(X['HouseFloor'] == 0)|(X['HouseFloor'] > 95), 'HouseFloor_outlier'] = 1
        X.loc[X['Floor'] > X['HouseFloor'], 'HouseFloor_outlier'] = 1
        
        X.loc[(X['HouseFloor'] == 0)|(X['HouseFloor'] > 95), 'HouseFloor'] = self.medians['HouseFloor']
        
        floor_outliers = X.loc[X['Floor'] > X['HouseFloor']].index
        X.loc[floor_outliers, 'Floor'] = X.loc[floor_outliers, 'HouseFloor']\
                                            .apply(lambda x: random.randint(1, x))
        
        # HouseYear
        current_year = datetime.now().year
        
        X['HouseYear_outlier'] = 0
        X.loc[X['HouseYear'] > current_year, 'HouseYear_outlier'] = 1
        
        X.loc[X['HouseYear'] > current_year, 'HouseYear'] = current_year
        
        # Healthcare_1
        X.loc[(X['Healthcare_1'].isna()), 'Healthcare_1'] = self.means['Healthcare_1']
            
        X['Id'] = X['Id'].astype(str)
        category_numbers = {"A": 0, "B": 1}
        X["Ecology_2"] = X["Ecology_2"].replace(category_numbers)
        X["Ecology_3"] = X["Ecology_3"].replace(category_numbers)
        X["Shops_2"] = X["Shops_2"].replace(category_numbers)     
        
        X.fillna(self.medians, inplace=True)
        
        return X
```

```python
class FeatureGenerator:
    def __init__(self):
        self.districts_median_year = None
        self.districts_density = None
        self.district_price_per_square = None
        self.min_sq = None
        self.max_sq = None

        

    
    def new_features(self, df):

        # Floor category
        floor_bins = [0, 4, 7, 12, df['Floor'].max()]
        df['Floor_cat'] = pd.cut(df['Floor'], bins=floor_bins, labels=False)
        df['Floor_cat'].fillna(-1, inplace=True) 
        
        
        # Scale and merge Social
        scaler = RobustScaler()
        pca = PCA(n_components=1, random_state=42)
        social_scaled = pd.DataFrame(scaler.fit_transform(df[['Social_1', 'Social_2', 'Social_3']]))
        df['Social'] = pca.fit_transform(social_scaled)
             
        df.drop(['Ecology_2', 'Ecology_3', 'Shops_2'], axis=1, inplace=True)
        
        return df
```

```python
preprocessor = DataPreprocessing()
preprocessor.fit(train_df)
features_gen = FeatureGenerator()
train_df = preprocessor.transform(train_df)
train_df = features_gen.new_features(train_df)
```

```python
%%time
params = {'n_estimators':[50, 100, 200, 400,1000,2000,4000], 
          'max_depth':[3, 5, 7, 10, 15,20,30]}

gs = GridSearchCV(gb_model, params, 
                  scoring='r2', 
                  cv=KFold(n_splits=5,   
                           random_state=42, 
                           shuffle=True),
                  n_jobs=-1
                  )
gs.fit(X_train, y_train)
```
