```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
DATASET_PATH = './creditcard.csv'
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv(DATASET_PATH, sep=',')
```

```python
plt_class_bar = df['Class'].value_counts().to_frame()
```

```python
plt_class_bar.plot(kind="bar")
plt_class_bar.plot(kind="bar",logy = True)
```

```python
colors = ['#ff0000', '#808080']
names = ['Class1', 'Class0']
plt.hist([df[(df.Class == 1)]['V1'],df[(df.Class == 0)]['V1']], bins = 20, density = True, color = colors, label=names)
plt.legend()
plt.xlabel('V1')
```
